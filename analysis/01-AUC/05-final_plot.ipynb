```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import roc_curve, roc_auc_score, auc
```

### ML results

#### 0. cs1 cs2

```python
tpr_cs1cs2 = np.loadtxt('rocs/roc_cs1cs2.txt')[0]
fpr_cs1cs2 = np.loadtxt('rocs/roc_cs1cs2.txt')[1]
roc_auc_cs1cs2 = auc(fpr_cs1cs2, tpr_cs1cs2)

tpr_cs1cs2ratio = np.loadtxt('rocs/roc_cs1cs2ratio.txt')[0]
fpr_cs1cs2ratio = np.loadtxt('rocs/roc_cs1cs2ratio.txt')[1]
roc_auc_cs1cs2ratio = auc(fpr_cs1cs2ratio, tpr_cs1cs2ratio)
```

#### 1. Compressed Features

```python
tpr_17feats = np.loadtxt('rocs/roc_17feats.txt')[0]
fpr_17feats = np.loadtxt('rocs/roc_17feats.txt')[1]
roc_auc_17feats = auc(fpr_17feats, tpr_17feats)

tpr_16feats = np.loadtxt('rocs/roc_16feats.txt')[0]
fpr_16feats = np.loadtxt('rocs/roc_16feats.txt')[1]
roc_auc_16feats = auc(fpr_16feats, tpr_16feats)

tpr_15feats = np.loadtxt('rocs/roc_15feats.txt')[0]
fpr_15feats = np.loadtxt('rocs/roc_15feats.txt')[1]
roc_auc_15feats = auc(fpr_15feats, tpr_15feats)

tpr_14feats = np.loadtxt('rocs/roc_14feats.txt')[0]
fpr_14feats = np.loadtxt('rocs/roc_14feats.txt')[1]
roc_auc_14feats = auc(fpr_14feats, tpr_14feats)

tpr_6feats = np.loadtxt('rocs/roc_6feats.txt')[0]
fpr_6feats = np.loadtxt('rocs/roc_6feats.txt')[1]
roc_auc_6feats = auc(fpr_6feats, tpr_6feats)

tpr_3featsv1 = np.loadtxt('rocs/roc_3featsv1.txt')[0]
fpr_3featsv1 = np.loadtxt('rocs/roc_3featsv1.txt')[1]
roc_auc_3featsv1 = auc(fpr_3featsv1, tpr_3featsv1)

tpr_3featsv2 = np.loadtxt('rocs/roc_3featsv2.txt')[0]
fpr_3featsv2 = np.loadtxt('rocs/roc_3featsv2.txt')[1]
roc_auc_3featsv2 = auc(fpr_3featsv2, tpr_3featsv2)
```

#### 3. Raw data as arrays

```python
tpr_arrays = np.loadtxt('rocs/roc_arrays.txt')[0]
fpr_arrays = np.loadtxt('rocs/roc_arrays.txt')[1]
roc_auc_arrays = auc(fpr_arrays, tpr_arrays)

tpr_arraysdt = np.loadtxt('rocs/roc_arraysdt.txt')[0]
fpr_arraysdt = np.loadtxt('rocs/roc_arraysdt.txt')[1]
roc_auc_arraysdt = auc(fpr_arraysdt, tpr_arraysdt)

tpr_arrays17feats = np.loadtxt('rocs/roc_arrays17feats.txt')[0]
fpr_arrays17feats = np.loadtxt('rocs/roc_arrays17feats.txt')[1]
roc_auc_arrays17feats = auc(fpr_arrays17feats, tpr_arrays17feats)
```

#### 4. Raw data as images

```python
tpr_images = np.loadtxt('rocs/roc_images.txt')[0]
fpr_images = np.loadtxt('rocs/roc_images.txt')[1]
roc_auc_images = auc(fpr_images, tpr_images)

tpr_imagesdt = np.loadtxt('rocs/roc_imagesdt.txt')[0]
fpr_imagesdt = np.loadtxt('rocs/roc_imagesdt.txt')[1]
roc_auc_imagesdt = auc(fpr_imagesdt, tpr_imagesdt)

tpr_images17feats = np.loadtxt('rocs/roc_images17feats.txt')[0]
fpr_images17feats = np.loadtxt('rocs/roc_images17feats.txt')[1]
roc_auc_images17feats = auc(fpr_images17feats, tpr_images17feats)
```

```python
lw=2

plt.plot(fpr_cs1cs2, tpr_cs1cs2, '-', color='coral', lw=lw, label="cS1, cS2, AUC = %0.3f" % roc_auc_cs1cs2)

# plt.plot(fpr_cs1cs2ratio, tpr_cs1cs2ratio, ':', color='red', lw=lw, label="cS1/cS2, AUC = %0.2f" % roc_auc_cs1cs2ratio)

plt.plot(fpr_16feats, tpr_16feats, '-', color='seagreen', lw=lw, label="16 resumed features, AUC = %0.3f" % roc_auc_16feats)
#plt.plot(fpr_17feats, tpr_17feats, '-', color='magenta', lw=lw, label="17 resumed features, AUC = %0.2f" % roc_auc_17feats)

# plt.plot(fpr_14feats, tpr_14feats, ':', color='green', lw=lw, label="14 resumed features, AUC = %0.3f" % roc_auc_14feats)
# # plt.plot(fpr_15feats, tpr_15feats, '-', color='magenta', lw=lw, label="15 resumed features, AUC = %0.2f" % roc_auc_15feats)

plt.plot(fpr_arrays, tpr_arrays, '-', color='magenta', lw=lw, label="Raw data (array), AUC = %0.3f" % 0.954)
#plt.plot(fpr_arrays17feats, tpr_arrays17feats, '-', color='darkorange', lw=lw, label="Raw data (array) + 17 features, AUC = %0.2f" % roc_auc_arrays17feats)

plt.plot(fpr_images, tpr_images, '-', color='dodgerblue', lw=lw, label="Raw data (images), AUC = %0.3f" % roc_auc_images)
#plt.plot(fpr_images17feats, tpr_images17feats, '-', color='blue', lw=lw, label="Raw data (images) + 17 features, AUC = %0.2f" % roc_auc_images17feats)


plt.xlabel("False Positive Rate",fontsize=14)
plt.ylabel("True Positive Rate",fontsize=14)
plt.xlim([-0.01,1.01])
plt.ylim([-0.01,1.01])
plt.yticks([0.2,0.4,0.6,0.8,1.])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.legend(loc="lower right", fontsize=12)
plt.savefig('rocs.pdf',bbox_inches='tight')
plt.show()
```
